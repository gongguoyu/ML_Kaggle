```python
def showColums():
    
    print('+'*20)
    
    print(data_train.columns.values.tolist())
    print('-'*20)
    print(data_test.columns.values.tolist())

    print('+'*20)
    
    data_train.info()
    print('-'*20)
    data_test.info()
    
    print('+'*20)
    
showColums()
```

```python
#无意义的标志位删除
def drop(ids):
    ids = list(set(ids))
    
    global data_train
    global data_test
    
    print('Drop:',ids)
    
    
    dealDatas = [data_train,data_test]
    dealDatas_Name = ['data_train','data_test']
    
    for i in range(2):
        print('-----',dealDatas_Name[i],':')
        
        for tempId in ids:
            try:
                dealDatas[i] = dealDatas[i].drop([tempId], axis=1)
            except Exception as e:
                print(e)
                pass
            else:
                print(tempId,)
                
    data_train,data_test = dealDatas

                
dropColumns = []      

dropColumns.append('Id')
drop(dropColumns)

showColums()
```

```python
#训练类别均值填充
def dofillNanUseMostType(_fillNanUseMostType):
    print('+'*20)
    print(data_train[_fillNanUseMostType].info())
    
    for name in _fillNanUseMostType:
        #print(data_train[name].value_counts())
        data_train[name][data_train[name].isna()]=data_train[name].mode().iloc[0]
        data_test[name][data_test[name].isna()]=data_train[name].mode().iloc[0]
        #print(data_train[name].value_counts())
        #print('-'*10)
        
    
    print(data_train[_fillNanUseMostType].info())
    print('+'*20)

dofillNanUseMostType(fillNanUseMostType)
```

```python
#未知类型填充
def dofillNanWithNan(_columnNeedDeal):
    print('+'*20)
    print(data_train[_columnNeedDeal].info())
    
    data_train[_columnNeedDeal] = data_train[_columnNeedDeal].fillna('NAN')
    data_test[_columnNeedDeal] = data_test[_columnNeedDeal].fillna('NAN')
    #for name in _columnNeedDeal:
    #    print(data_train[name].value_counts())
    #    print('-'*10)
    
    print(data_train[_columnNeedDeal].info())
    print('+'*20)

dofillNanWithNan(fillNanWithNan)
```

```python
#类别均值填充
def dofillNanUseMostTypeMean(_data):
    
    _columnNeedDeal = [x[0] for x in _data]
    _columnContainMessage = [x[1] for x in _data]
    
    print(_columnNeedDeal)
    print(_columnContainMessage)
    
    print('+'*20)
    print(data_train[_columnNeedDeal].info())
    
    for c_deal,c_message in zip(_columnNeedDeal,_columnContainMessage):
        #print(data_train[m][data_train[c].isna()])
        ids = data_train[c_deal].isna()
        ids_test = data_test[c_deal].isna()
        
        #print(data_train[c_deal][ids])
        
        gr = data_train.groupby([c_message])[c_deal].mean()
        data_train[c_deal][ids] = data_train[c_message][ids].apply(lambda x:gr[x])
        data_test[c_deal][ids_test] = data_test[c_message][ids_test].apply(lambda x:gr[x])
        
        #print(data_train[c_deal][ids])
        print('-'*10)
    
    print(data_train[_columnNeedDeal].info())
    print('+'*20)

dofillNanUseMostTypeMean(fillNanUseMostTypeMean)
```

```python
#使用其他特征填充
def dofillNanWithOtherFeature(_data):
    
    _columnNeedDeal = [x[0] for x in _data]
    _columnContainMessage = [x[1] for x in _data]
    
    print(_columnNeedDeal)
    print(_columnContainMessage)
    
    print('+'*20)
    print(data_train[_columnNeedDeal].info())
    
    for c_deal,c_message in zip(_columnNeedDeal,_columnContainMessage):
        ids = data_train[c_deal].isna()
        data_train[c_deal][ids]=data_train[c_message][ids]
        
        ids_test = data_test[c_deal].isna()
        data_test[c_deal][ids_test]=data_test[c_message][ids_test]
    
    print(data_train[_columnNeedDeal].info())
    print('+'*20)

dofillNanWithOtherFeature(fillNanWithOtherFeature)
```
